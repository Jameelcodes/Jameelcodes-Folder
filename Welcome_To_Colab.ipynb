```
import pandas as pd
import geopandas as gpd
```

```
# Load flood data from CSV
flood_data = pd.read_csv('/content/2023 (1).csv') # Use pd.read_csv() to load the data
print(flood_data.head())
```

```
# Load weather data from CSV
weather_data = pd.read_csv('/content/2023 (1).csv') # Use pd.read_csv() to load the data
print(weather_data.head())
```

```
# Load topographical data from a shapefile or GeoJSON
topo_data = gpd.read_file('/content/2023 (1).csv') # Use gpd.read_file() to load geospatial data
print(topo_data.head())
```

```
# Load urbanization data from CSV
urban_data = pd.read_csv('/content/2023 (1).csv') # Use pd.read_csv() to load the data
print(urban_data.head())
```

```
# Convert datetime columns to datetime
flood_data['datetime'] = pd.to_datetime(flood_data['datetime'])
weather_data['datetime'] = pd.to_datetime(weather_data['datetime'])
urban_data['datetime'] = pd.to_datetime(urban_data['datetime'])
```

```
# Filter data to the same time range
start_datetime = '2023-01-01'
end_datetime = '2023-12-31'
flood_data = flood_data[(flood_data['datetime'] >= start_datetime) & (flood_data['datetime'] <= end_datetime)]
weather_data = weather_data[(weather_data['datetime'] >= start_datetime) & (weather_data['datetime'] <= end_datetime)]
urban_data = urban_data[(urban_data['datetime'] >= start_datetime) & (urban_data['datetime'] <= end_datetime)]
```

```
# Impute missing values for numeric columns only
numeric_columns = weather_data.select_dtypes(include=['number']).columns
weather_data[numeric_columns] = weather_data[numeric_columns].fillna(weather_data[numeric_columns].mean())

urban_data.fillna(method='ffill', inplace=True)  # Forward fill for urban data
```

```
# Calculate cumulative precip
weather_data['cumulative_precip'] = weather_data['precip'].cumsum()

# Calculate days since last flood
flood_data['days_since_last_flood'] = (flood_data['datetime'] - flood_data['datetime'].shift()).dt.days
```

```
# Summary statistics
print(flood_data.describe())
print(weather_data.describe())
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Histogram of precip
sns.histplot(weather_data['precip'])
plt.title('Precip Distribution')
plt.show()
```

```
# Box plot of flood occurrences
sns.boxplot(x='datetime', y='flood_count', data=flood_data)
plt.title('Yearly Flood Occurrences')
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Define the data
data = {
    'Location': [
        'Akabo Hall I, Nnewi South, Anambra',
        'Nkwo Amenyi Square II, Akwa South, Anambra',
        'Akwa Village Hall III, Dunukofia, Anambra'
         ],
    'Votes': [350, 228, 148],
    'Cluster': [0, 0, 0],
    'Outlier Score': [345.6804124, 226.1570881, 144.8373984]
}
```

```
# Create a DataFrame
df = pd.DataFrame(data)
```

```
# Set the plot style
sns.set(style="whitegrid")
```

```
# Create a bar plot for outlier scores
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Outlier Score', y='Location', data=df, palette='viridis')
```

```
# Add title and labels
plt.title('Top 3 Outliers by Outlier Score')
plt.xlabel('Outlier Score')
plt.ylabel('Location')
```

```
# Add the outlier scores on the bars
for index, value in enumerate(df['Outlier Score']):
    barplot.text(value, index, f'{value:.2f}', color='black', ha="left", va="center")
```

```
# Adjust layout to ensure everything fits without overlap
plt.tight_layout()
```

```
# Display the plot
plt.show()
```
